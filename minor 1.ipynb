# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
data = pd.read_csv('Churn_Modelling.csv')
data
```

## Basic Understand of Data

```python
data.shape
data.info()
data.describe()

# why used: To understand number of rows, columns, data types, and basic statistics.
```

## Data Cleaning

```python
data.isnull().sum()  

#why used: check missing value
```

```python
data.drop(columns=['RowNumber','Surname'],inplace=True)

#why used : To drop unnecessary columns
```

# Univariate analysis :
Analysis of a single variable to understand its distribution and behavior.

## count plot

```python
data['Exited'].value_counts()
sns.countplot(x='Exited',data=data,hue='Gender',palette='Accent')
plt.title('Customer Churn Comparison Based on Gender')
plt.grid()

# This countplot shows the distribution of customer churn(Exited) across different genders.
```

## pie chart

```python
data['Exited'].value_counts().plot(kind='pie',autopct='%2f',colors=['#4CAF50', '#F44336'])
plt.title('Overall Customer Churn Percentage Distribution')
# Visualizes overall customer churn distribution in percentage form.
```

# Bivariate Analysis:
Analysis of two variables to understand the relationship between them.

```python
sns.countplot(x='Geography', hue='Exited', data=data)
plt.title('Customer Churn Distribution Across Geographical Regions') 
plt.show()

## This countplot shows how many customers exited vs not exited across different geographical regions.
```

## Multivariate Analysis :
Analysis of more than two variables at the same time to study their combined effect.

## scatter plot

```python
sns.scatterplot(x='Age',y='Balance',data = data,hue='Exited')
plt.title("Relationship Between Customer Age and Balance by Churn")
plt.show()

# Display how customer balance varies age for churned and non-churned customers.
```

## boxplot

```python
sns.boxplot(x='Exited',y='Age',data=data,hue='Gender')
plt.title('Age Distribution of Customers by Churn Status and Gender')

# This boxplot campares the age distribution of customers who exited and did not exit, separated by gender.
```

# displot

```python
sns.displot(data['Age'], kind='kde')
plt.title('Age Distribution')
plt.show()
data['Age'].skew()

#This distribution plot visualizes the spread and skewness of the Age variable to understand its shape before modeling.
```
